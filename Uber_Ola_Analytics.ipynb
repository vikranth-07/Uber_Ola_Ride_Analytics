```python
import pandas as pd

df = pd.read_csv("uber.csv")

df.rename(columns={
    'Date/Time': 'pickup_datetime',
    'Lat': 'pickup_latitude',
    'Lon': 'pickup_longitude',
    'Base': 'uber_base'
}, inplace=True)

df.head()
```

```python
import numpy as np

# Estimated trip distance
df['trip_distance'] = np.random.uniform(1, 20, len(df))

# Estimated fare
df['fare_amount'] = df['trip_distance'] * 25

# Driver ratings
df['driver_rating'] = np.random.uniform(3.5, 5.0, len(df))

# Vehicle type
df['vehicle_type'] = np.random.choice(
    ['Mini', 'Sedan', 'SUV'],
    len(df)
)

# Ride status
df['ride_status'] = np.random.choice(
    ['Completed', 'Cancelled'],
    len(df),
    p=[0.9, 0.1]
)

# Payment method
df['payment_method'] = np.random.choice(
    ['Cash', 'Card', 'UPI'],
    len(df)
)
```

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])

df['hour'] = df['pickup_datetime'].dt.hour
df['day'] = df['pickup_datetime'].dt.day_name()
df['month'] = df['pickup_datetime'].dt.month_name()
```
